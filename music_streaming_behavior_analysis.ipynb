```python
# Converting all column names in the DataFrame to lowercase
df.columns = [column_name.strip().lower() for column_name in df.columns]
df = df.rename(columns={"userid": "user_id"})
```

```python
df.isna().sum()
```

```python
# Going through the headers and replacing the empty spaces with 'unknown'
cols = ['track','artist','genre']
df[cols] = df[cols].fillna('unknown')
```

```python
df.isna().sum(0)
```

```python
df.duplicated().sum()
```

```python
# removing duplicates and resetting the index
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.duplicated().sum()
```

```python
col_genre = df['genre']
print(sorted(col_genre.unique()))
```

```python
def replace_wrong_genres(wrong_genres, correct_genre):
    for wrong_genre in wrong_genres:
        df['genre'] = df['genre'].replace(wrong_genre, correct_genre)
```

```python
wrong_genres = ['hip', 'hop', 'hip-hop']
correct_genre = 'hiphop'

replace_wrong_genres(wrong_genres, correct_genre)
```

```python
print(sorted(df['genre'].unique()))
```

```python
# Counting the songs played in each city
plays_by_city = df.groupby('city')['track'].count().sort_values(ascending=False)
print(plays_by_city)
```

```python
# Counting the songs played on each day of the week
df_days = df[(df['day'] == 'Monday') | (df['day'] == 'Wednesday') | (df['day'] == 'Friday')]
```

```python
for days in ['Monday', 'Wednesday', 'Friday']:
    print(f"- {days}s:")                       # cabeçalho simplificado
    print(df[df['day'] == days].groupby('city')['track']
          .count()
          .sort_values(ascending=False),"\n")
```
